# Part I: Three observable trends based on the WeatherPy data


The plot of Northern Hemisphere Maximum Temperatures (*C), above, shows a strong relationship with respect to Latitude. The high r-squared value and tight fit of the data points to the regression line suggest that there is a strong negative correlation between these two variables. This is intuitive, since as we increase in Latitude, moving towards the North Pole, the maximum temperature of a location there typically decreases. It gets quite a bit cooler up there vs. the equator!



Like the Northern Hemisphere, the Southern Hemisphere Maximum Temperatures (*C) plot, above, also shows a strong relationship with Latitude. The relatively high r-squared value and tight fit of the linear regression plot suggest that there is a string positive correlation between these two variables (positive this time because as latitiude increases in the Southern Hemisphere we're approaching the equator instead of leaving it, as in the North). 



In this third plot, above, we have Global Humidity (%) plotted against Latitude. Visually, it seems like the data is fairly evenly distributed without much of a pattern - although interestingly, it does seem like the regions of the world's deserts may be visible in this representation of the data, clustered around -20* and between 20* to 40* of latitude.  

In the chart below, the linear regression and relatively low r-squared value (0.07459) of the Northern Hemisphere's Humidity (like the South's) shows that there's little correlation between Humidity and Latitude.
